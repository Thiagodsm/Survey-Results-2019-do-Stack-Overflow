**How many people from Brazil knows Python**(first: using filter, and then using groupby)

PS: people that answer the Stack Overflow survey's

```python
filt = df['Country'] == 'Brazil'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) #.value_counts(normalize=True)
```

```python
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).loc['Brazil']
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

**Number of people of each country that answered the survey and the number that knows Python**


```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith':'NumKnowsPython'}, inplace=True)
python_df
```

**Now We can calculate the percentual of each country how knows Python**

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df.loc['Brazil']
```
